2. Load the dataset into a pandas DataFrame stocke it in a df variable.

```python
df = pd.read_csv("pokemon_data.csv")
```

4. Show its shape, data types, and count of missing values per column.

```python
# The shape:
print(df.shape)
print("--------")
# The dataset has 800 rows and 12 column.
# The data types of column:
print(df.dtypes)
print("--------")
# The missing values:
print(df.info())
# The table present 414 non-null in 'Type 2', means 800 - 414 = 386 missing values
print("--------")
# or
print(df.isna().sum())
```

5. Select all Pokémon whose Type 1 is “Fire” and whose Attack stat is greater than 70.

```python
print(df[(df["Type 1"] == "Fire") & (df["Attack"] > 70)].head(3))
# or
import numpy as np
bool_data = np.logical_and(df["Type 1"] == "Fire", df["Attack"] > 70)
print(df[bool_data].head(1))
```

6. Sort the DataFrame by HP in descending order and then by Speed in ascending order. Display the top 5.

```python
data1 = df.sort_values("HP", ascending=False)
print(data1.sort_values("Speed").head(5))
```

7. Create a new column called TotalStats which is the sum of HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed.

```python
df["TotalStats"] = df["Attack"]+df["HP"]+df["Defense"]+df["Speed"]+df["Sp. Atk"]+df["Sp. Def"]
# or 
new_dataset = df.assign(TotalStats=lambda x: x.iloc[:, 4:10].sum(axis=1))
new_dataset.sample(3)
```

8. Find the average TotalStats for each Type 1. Which type has the highest average?

```python
all_type1 = df.groupby(["Type 1"])["TotalStats"].mean()
print(f"The Type that has the highest average '{all_type1.max()}' is: {all_type1.idxmax()}")
```

9. Filter the dataset to keep only Pokémon of generation 1 or 2, and then drop duplicates based on the Name column (keeping the one with the highest TotalStats).

```python
gener1_2_data = df[(df["Generation"]==1) | (df["Generation"]==2)]
# or
bool_gener1_2_data = np.logical_or(df["Generation"]==1, df["Generation"]==2)
df[bool_gener1_2_data].head()
# drop duplicates:
gener1_2_data.drop_duplicates("Name").head()
```

10. For Pokémon whose Legendary flag is true, compute how many there are, and show the top 3 highest TotalStats among them.

```python
# Nembers of legendary pokemons:
legendary_data = df[df["Legendary"]]
nbr_of_legendary = legendary_data["Legendary"].count()
print(nbr_of_legendary)
# The top 3 highest TotalStats:
print(legendary_data.sort_values("TotalStats").head(3))
```

11. Convert the Name column to lowercase.

```python
df["Name"] = df["Name"].str.lower()
```

12. Find the strongest Pokémon in each Type 1 + Type 2 combination.

```python
strongest_pok = df.groupby(["Type 1", "Type 2"])["TotalStats"].mean()
f"The strongest pokemone has {strongest_pok.max()} in TotalStats is: {strongest_pok.idxmax()}"
```

13. Find all Pokémon with unusually high Speed.

```python
unusuallt_speed = df[df["Speed"] > df["Speed"].mean() + 2 * df["Speed"].std()]
unusuallt_speed[["Name", "Type 1", "Generation", "Speed"]].head()
```

14. Find Pokémon whose name contains:
- The substring “mega”
- Ends with “ny”
- Has exactly 10 characters
Return the counts for each category.

```python
# The pokemons has the substring "mega":
has_mega = df["Name"][df["Name"].str.contains("Mega")].count()
# Ends with "ny":
end_by_ny = df["Name"][df["Name"].str.endswith("ny")].count()
# 10 characters:
teen_char = df["Name"][df["Name"].str.len()==10].count()
print(f"The {has_mega} have the substring 'Mega' \n{end_by_ny}: ends with 'ny'\n{teen_char}: has exactly 10 characters")
```

15. Compare average stats of Legendary vs Non-Legendary.

```python
df.groupby(["Legendary"])["TotalStats"].mean(numeric_only=True)
```

16. Find all Type 1 categories where median Attack > 75.

```python
group_type1 = df.groupby("Type 1").mean(numeric_only=True)
group_type1[["Attack"]][group_type1["Attack"] > 75]
```
